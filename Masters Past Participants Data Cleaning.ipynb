# Data Cleaning 

Here we will Clean the Mster's data that we have collected so far and format it into what we need for statistical analysis

Loading in our packages:

```python
import pandas as pd
import numpy as np
import scipy
```

```python
masters_df = pd.read_csv("C:\Personal Projects\masters_players.csv")

masters_df.head(10)
```

Mutating Pos column (We need to change our ties so that we drop the "T")

```python
masters_df["Pos"] = masters_df["Pos"].str.replace(r"^T", "", regex = True)

masters_df["Pos"].head(30)
```

Now We need to create another column "Made Cut" based on whether or not a player made the cut that respective year 

```python
masters_df["Made_Cut"] = np.where(masters_df["Pos"] == "MC", 0, 1)

masters_df["Made_Cut"].head(70)
```

```python
masters_df.head(20)
```

Let's also mutate everyone who missed cuts or round scores say to illness/injury, dq's etc to NA's for data analysis

```python
masters_df["R1"] = masters_df["R1"].replace("", np.nan)
masters_df["R2"] = masters_df["R2"].replace("", np.nan)
masters_df["R3"] = masters_df["R3"].replace("", np.nan)
masters_df["R4"] = masters_df["R4"].replace("", np.nan)
masters_df["Total Score"] = masters_df["Total Score"].replace("", np.nan)
masters_df["Total Par"] = masters_df["Total Par"].replace("", np.nan)
```

Now let's start changing data types: we know the year should be a date, the names should be string, and all of the scores should be numeric

```python
masters_df["Pos"] = masters_df["Pos"].astype("category")

masters_df["Year"] = masters_df["Year"].astype(int)

masters_df["Name"] = masters_df["Name"].astype("string")

masters_df["R1"] = pd.to_numeric(masters_df["R1"])

masters_df["R2"] = pd.to_numeric(masters_df["R2"])

masters_df["R3"] = pd.to_numeric(masters_df["R3"])

masters_df["R4"] = pd.to_numeric(masters_df["R4"])

masters_df["Total Score"] = pd.to_numeric(masters_df["Total Score"])

masters_df["Total Par"] = (masters_df["Total Par"].replace("E", 0))

masters_df["Total Par"] = pd.to_numeric(masters_df["Total Par"])

masters_df["Made_Cut"] = pd.to_numeric(masters_df["Made_Cut"])


masters_df.info()
```

Now let's output the csv as our finalized masters participants 

```python
masters_df.to_csv("C:/Personal Projects/finalized_masters_participants.csv", index=False)
```
